```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)
%matplotlib inline

X = np.linspace(0, 100, 100)

# Compute X squared
Y = X**2

# Setup figure and axes
fig, ax = plt.subplots()

# Plot X versus X^2
ax.plot(X, Y, label="X Squared", color="blue")

# Customize the plot
ax.set_title("Plot of X vs X^2")
ax.set_xlabel("X values")
ax.set_ylabel("X Squared values")
ax.legend()  # Add lege
plt.show()

```

We'll start with scatter plots.

```python
# Create a scatter plot of X versus the exponential of X (np.exp(X))
X = np.linspace(0, 5, 100)

Y = np.exp(X)

fig, ax = plt.subplots()

ax.scatter(X, Y, color="red", label="exp(X)", alpha=0.7)

# Customize the plot
ax.set_title("Scatter Plot of X vs exp(X)")
ax.set_xlabel("X values")
ax.set_ylabel("Exponential of X")
ax.legend()  # Add legend

# Show the plot
plt.show()
```

```python
# Create a scatter plot of X versus np.sin(X)

X = np.linspace(0, 10, 100)

# Compute the sine of X
Y = np.sin(X)

# Setup figure and axes
fig, ax = plt.subplots()

# Create a scatter plot of X versus sin(X)
ax.scatter(X, Y, color="blue", label="sin(X)", alpha=0.7)

# Customize the plot
ax.set_title("Scatter Plot of X vs sin(X)")
ax.set_xlabel("X values")
ax.set_ylabel("sin(X)")
ax.legend()  # Add legend

# Show the plot
plt.show()
```

Now let's try make some subplots. A subplot is another name for a figure with multiple plots on it.

```python
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns (4 subplots total)
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Set titles for each subplot (optional)
axes[0, 0].set_title("Subplot 1")
axes[0, 1].set_title("Subplot 2")
axes[1, 0].set_title("Subplot 3")
axes[1, 1].set_title("Subplot 4")

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the empty subplots
plt.show()
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
X = np.linspace(0, 10, 100)  # 100 evenly spaced numbers between 0 and 10
Y = X / 2  # X divided by 2

random_x = np.random.rand(10)  # 10 random numbers for x-axis
random_y = np.random.rand(10)  # 10 random numbers for y-axis

favorite_foods = {"Pizza": 10.99, "Burger": 5.49, "Ice Cream": 3.99}
food_names = list(favorite_foods.keys())
food_prices = list(favorite_foods.values())

random_data = np.random.randn(1000)  # 1000 normally distributed numbers

# Create a 2x2 grid of subplots with figsize (10, 5)
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Top-left: Line plot of X vs X/2
axes[0, 0].plot(X, Y, color="blue", label="X/2")
axes[0, 0].set_title("X vs X/2")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")
axes[0, 0].legend()

# Top-right: Scatter plot of random points
axes[0, 1].scatter(random_x, random_y, color="red", label="Random Points")
axes[0, 1].set_title("Random Scatter Plot")
axes[0, 1].set_xlabel("Random X")
axes[0, 1].set_ylabel("Random Y")
axes[0, 1].legend()

# Bottom-left: Bar chart of favorite foods
axes[1, 0].bar(food_names, food_prices, color=["red", "green", "blue"])
axes[1, 0].set_title("Favorite Foods and Prices")
axes[1, 0].set_xlabel("Food Items")
axes[1, 0].set_ylabel("Price ($)")

# Bottom-right: Histogram of normal distribution
axes[1, 1].hist(random_data, bins=30, color="purple", edgecolor="black", alpha=0.7)
axes[1, 1].set_title("Histogram of Normal Distribution")
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Frequency")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Import pandas as pd
import pandas as pd
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view

car_sales = pd.read_csv("car-sales.csv")

# Display the first 5 rows of the DataFrame
car_sales.head()
```

```python
# Try to plot the 'Price' column using the plot() function

# Remove non-numeric characters and convert 'Price' to float
car_sales["Price"] = car_sales["Price"].replace('[\$,]', '', regex=True).astype(float)

car_sales["Price"].plot()

# Customize the plot
plt.title("Car Prices Over Index")
plt.xlabel("Index")
plt.ylabel("Price ($)")
plt.show()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers

print(car_sales["Price"].head())  # Display the first 5 cleaned values
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column (cumsum())
# Add 'Total Sales' column (cumulative sum of 'Price')
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Add 'Sale Date' column (generate successive dates starting from today)
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales), freq='D')

# View the updated DataFrame
(car_sales.head())
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(car_sales["Sale Date"], car_sales["Total Sales"], marker='o', linestyle='-')

plt.show()
```

```python
# Convert the 'Price' column to the integers
car_sales["Price"] = car_sales["Price"].astype(int)
print(car_sales["Price"].dtype) 
```

```python
# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function

plt.figure(figsize=(8, 5))

plt.scatter(car_sales["Odometer (KM)"], car_sales["Price"], alpha=0.7)

plt.title("Car Price vs. Odometer (KM)")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price ($)")
plt.grid(True)


plt.show()
```
